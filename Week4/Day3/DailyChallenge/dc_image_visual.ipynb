```python
# Load and Visualize Images using the Flower Color Images dataset
# Rotate an image by 90 degrees
# Flip an image horizontally and then vertically.
# Zoom in on an image (scale by 1.2x) using .resize(...)
# Display the original and augmented images side by side for comparison using this function :
```

```python
import matplotlib.pyplot as plt
from PIL import Image, ImageOps

def show_images(original, augmented, title):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(augmented)
    plt.title(title)
    plt.axis('off')

    plt.show()
```

```python
image = Image.open('0002.png')

rotated_image = image.rotate(90)
flipped_horizontally = image.transpose(Image.FLIP_LEFT_RIGHT)
flipped_vertically = image.transpose(Image.FLIP_TOP_BOTTOM)
k = 1.2
zoomed_image = image.resize((int(image.width * k), int(image.height * k)))
```

```python
show_images(image, rotated_image, 'Rotated Image')
```

```python
show_images(image, flipped_horizontally, 'Flipped Horizontally')
```

```python
show_images(image, flipped_vertically, 'Flipped Vertically')
```

```python
show_images(image, zoomed_image, 'Zoomed Image')
```
