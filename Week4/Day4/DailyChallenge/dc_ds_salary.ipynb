```python
import pandas as pd

df = pd.read_csv('datascience_salaries.csv', index_col=0)

df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

scaler = MinMaxScaler()
df['salary_normalized'] = scaler.fit_transform(df[['salary']])

df.describe()
```

```python
ohe = pd.get_dummies(df[['job_title', 'job_type', 'experience_level', 'salary_currency']], drop_first=False)
df = pd.concat([df, ohe], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['location_label'] = le.fit_transform(df['location'])
```

```python
df.info()
```

```python
df.drop('salary', axis=1).head()
```

```python
pca = PCA(n_components=2)
pca_components = pca.fit_transform(df.select_dtypes(include=[float, int]).drop('salary', axis=1))

df['PCA1'] = pca_components[:, 0]
df['PCA2'] = pca_components[:, 1]

df.head()
```
