## Exercise 2 : Creating a Line Plot for Temperature Variation Instructions

- Create a simple line plot using Matplotlib that represents the temperature variations over a week.
- Use a list of temperature values for each day of the week
(e.g., [72, 74, 76, 80, 82, 78, 75]) and label the x-axis as “Day” and the y-axis as “Temperature (°F)”.

```python
import matplotlib.pyplot as plt

temperatures = [72, 74, 76, 80, 82, 78, 75]
days = range(1, 8)

plt.plot(days, temperatures, marker='.')
plt.grid(True)
plt.xticks(days)
plt.xlabel('Day')
plt.ylabel('Temperature (°F)')
plt.title('Temperature Over Days')

plt.show()
```

## Exercise 3 : Visualizing Monthly Sales with a Bar Chart

- Generate a bar chart using Matplotlib to visualize the monthly sales data for a retail store.
- Create a list of sales values for each month (e.g., [5000, 5500, 6200, 7000, 7500]) and label the x-axis as “Month” and the y-axis as “Sales Amount ($)”.

```python
import numpy as np
import datetime
import pandas as pd
import matplotlib.dates as mdates

n = 15

sales = np.random.randint(5000, 7500, size=n)
months = pd.date_range('2014-10-10', periods=n, freq='MS')

plt.figure(figsize=(10, 5))
plt.bar(months, sales, width=20)  
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales Data')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

plt.gcf().autofmt_xdate()

for i, v in enumerate(sales):
    plt.text(months[i], v, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()

```

## 🌟 Exercise 4 : Data Visualisation
- Calculate the total quantity of products sold by the company.
- Identify the category that had the highest revenue and calculate how much revenue it generated.
- Determine the average revenue per sale in the dataset.
- Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).


```python
import pandas as pd

df = pd.read_csv('sales-data-analysis/sales_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum(), df.nunique()
```

```python
df.columns 
```

```python
# - Calculate the total quantity of products sold by the company.
print('Total quantity sold: ', df.quantity.sum().astype(int))
```

```python
df.category.value_counts()
```

```python
df.category.replace('Bgas', 'Bags', inplace=True)
df.category.replace('Clohting', 'Clothing', inplace=True)
df.category.replace('Shoeses', 'Shoes', inplace=True)
```

```python
# - Identify the category that had the highest revenue and calculate how much revenue it generated.
category_revenue = df.groupby('category')['revenue'].sum().sort_values(ascending=False)
category_revenue
```

```python
print('Category with the highest revenue: ', category_revenue.index[0], category_revenue.values[0])
```

```python
import matplotlib.pyplot as plt

# Plot category revenue using pandas tools
plt.figure(figsize=(10, 6))
category_revenue.plot.bar()
plt.title('Total Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```

```python
# - Determine the average revenue per sale in the dataset.
print('Average revenue per sale: ', df.revenue.mean().round(2))
```

```python
df.columns
```

```python
# - Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).

df['quarter'] = pd.to_datetime(df['date']).dt.quarter
quarterly_revenue = df.groupby('quarter')['revenue'].sum()
print(quarterly_revenue)

plt.figure(figsize=(10, 6))
quarterly_revenue.plot.bar()
plt.title('Total Revenue by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Revenue')
plt.xticks(quarterly_revenue.index - 1, ['Q1', 'Q2', 'Q3', 'Q4'])
plt.xticks(rotation=0)

plt.show()
```

## 🌟 Exercise 5 : Data Visualisation Using MatPlotLib

- Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.
- Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.
- Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.

```python

# Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. 
# Add a title and label the x and y axes.
import numpy as np

x = np.linspace(-10, 10, 100)
y = x**2

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = x^2')
plt.title('Plot simple curve')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['$y = x^2$'], loc='upper right')

plt.show()


```

```python

# Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week.
# The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.

products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]

plt.figure(figsize=(10, 6))
plt.bar(products, sales, color=plt.cm.viridis(np.linspace(0, 1, len(products))))
plt.title('Sales Data of Four Products')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python

# Create a pie chart representing the following data about favorite fruits of a group of people:
# Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). 
# Customize the chart with different colors for each fruit and add a legend.

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [40, 30, 20, 10]

plt.figure(figsize=(10, 6))
plt.pie(sales, labels=fruits, autopct='%.0f%%', startangle=90,
        colors=['green', 'yellow', 'red', 'brown'])
plt.title('Favorite Fruits of a Group of People')
plt.legend(fruits, loc='upper right')

plt.show()
```
