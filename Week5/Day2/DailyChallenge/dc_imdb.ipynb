```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('datasets/imdb_top_1000.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
df.info()
```

```python
def is_numeric(col):
    try:
        pd.to_numeric(col)
        return True
    except:
        return False
```

```python
df['Released_Year'][~df['Released_Year'].apply(is_numeric)]
```

```python
df.loc[966, 'Released_Year'] = 1995 
```

```python
# Let's process some columns

df['Released_Year'] = df['Released_Year'].astype(int)
df['Runtime'] = df['Runtime'].str.replace(' min', '').astype(int)
df['Gross'] = df['Gross'].str.replace(',', '').astype(float)
```

```python
# Next lets handle missing values

df.isnull().sum()
```

```python
ms = df['Meta_score'].dropna()

plt.figure(figsize=(10, 6))
sns.histplot(ms, kde=True)

plt.axvline(ms.mean(), color='r', linestyle='--', label='mean')
plt.axvline(ms.median(), color='g', linestyle='--', label='meadian')

plt.legend()

plt.show()

print(f"mean: {ms.mean():.2f}")
print(f"median: {ms.median():.2f}")
```

```python
# mean and median are pretty close so lets fill missing values with mean

df['Meta_score'] = df['Meta_score'].fillna(df['Meta_score'].mean())
```

```python
gr = df['Gross'].dropna()

plt.figure(figsize=(10, 6))
sns.histplot(gr, kde=True)

plt.axvline(gr.mean(), color='r', linestyle='--', label='mean')
plt.axvline(gr.median(), color='g', linestyle='--', label='meadian')

plt.legend()

plt.show()

print(f"mean: {gr.mean():.2f}")
print(f"median: {gr.median():.2f}")
```

```python
# distribution is right skewed so it is better to use median for filling missing values
df['Gross'] = df['Gross'].fillna(df['Gross'].median())
```

Rating Trends Over Years:



```python
rating_trends = df.groupby('Released_Year')['IMDB_Rating'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.regplot(x='Released_Year', y='IMDB_Rating', data=rating_trends, 
            scatter_kws={'s': 50}, 
            line_kws={'color': 'red'})

sns.lineplot(x='Released_Year', y='IMDB_Rating', data=rating_trends)

plt.title('IMDb Rating Trends Over the Years')
plt.grid(True)
plt.show()
```

```python
# It seems that movie ratings have a slite downtrend and also it looks like variation lowers over time  
```

Genre Popularity Analysis:



```python
df['Genre_list'] = df['Genre'].str.split(', ')
```

```python
df[['Genre_list']].explode('Genre_list').value_counts()
```

```python
genre_counts = df['Genre_list'].explode().value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.title('Number of Movies by Genre', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

Director’s Impact on Ratings:



```python
df['Director'].nunique()
```

```python
#  lets pick top 20 (by avg rating) directors because where are too many of them
director_ratings = df.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending=False).reset_index().iloc[:20]

top_director_data = df.merge(director_ratings, on='Director', suffixes=('', '_mean'), how='inner').sort_values('IMDB_Rating_mean', ascending=False)

plt.figure(figsize=(12,6))
sns.scatterplot(data=top_director_data, x='Director', y='IMDB_Rating', alpha=0.7)

plt.title('Impact of The Directors on IMDb Ratings', fontsize=16)
plt.xticks(rotation=90)
plt.show()
```

```python
# it seems that this dataset doesnt have exhaustive list of films of every director so
# it would be better to include only those who has at least 3 films  
```

```python
#  lets pick top 20 (by avg rating) directors because where are too many of them
directors_3_plus = df['Director'].value_counts()
directors_3_plus = directors_3_plus[directors_3_plus.values >= 3]

director_ratings = df[df['Director'].isin(directors_3_plus.index)]
director_ratings = director_ratings.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending=False).reset_index().iloc[:20]

top_director_data = df.merge(director_ratings, on='Director', suffixes=('', '_mean'), how='inner').sort_values('IMDB_Rating_mean', ascending=False)

plt.figure(figsize=(15,6))
sns.scatterplot(data=top_director_data, x='Director', y='IMDB_Rating', alpha=0.7)
sns.lineplot(data=top_director_data, x='Director', y='IMDB_Rating_mean', color='red', linewidth=2)

plt.title('Impact of The Directors on IMDb Ratings', fontsize=16)
plt.xticks(rotation=90)
plt.ylabel('IMDb Rating')
plt.legend(['Individual Movies', 'Average Rating'])
plt.show()
```

Star Power Analysis:


```python
df.columns
```

```python
df[df['Series_Title'].duplicated(keep=False)]
```

```python
# there is duplicates by title so let add index as id for cinvinience 
df['film_id'] = df.index.copy()
```

```python

stars = df[['Star1', 'Star2', 'Star3', 'Star4', 'IMDB_Rating', 'Series_Title', 'Gross', 'film_id']].melt(
    ['IMDB_Rating', 'Series_Title', 'Gross', 'film_id'], var_name='star_num', value_name='star_name')
stars.head()
```

```python
import numpy as np
# For simplicity we calculate stars power as avg rating of every film they took part in 
star_power = stars.groupby('star_name')[['IMDB_Rating', 'Gross', 'film_id']].agg({'IMDB_Rating': 'mean', 'Gross': 'median', 'film_id': 'count'}).add_prefix('star_')
star_power
```

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(data=star_power, x='star_film_id')
plt.title('Distribution of Number of Films per Star')

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', 
                xytext=(0, 5), textcoords='offset points')

plt.xlabel('Film Count')
plt.show()
```

```python
stars = stars.merge(star_power, on='star_name')
stars.head()
```

```python
stars.columns
```

```python
df = df.merge(stars.groupby('film_id')[['star_IMDB_Rating', 'star_Gross']].agg({'star_IMDB_Rating': 'mean', 'star_Gross': 'median'}),
              left_index=True, right_index=True)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Select relevant columns for the pairplot
columns_of_interest = ['IMDB_Rating', 'Gross', 'star_IMDB_Rating', 'star_Gross']

# Create a copy of the dataframe with log-scaled Gross columns
df_log = df[columns_of_interest].copy()
df_log['Gross'] = np.log10(df_log['Gross'])
df_log['star_Gross'] = np.log10(df_log['star_Gross'])

# Create the pairplot with log-scaled Gross columns
g = sns.pairplot(df_log, height=2.5, aspect=1.2, 
                 plot_kws=dict(alpha=0.4, edgecolor="none", s=10))

# Customize the plot
g.figure.suptitle("Relationships between Film Ratings, Log Gross, and Star Power")
plt.tight_layout()
plt.show()
```

```python
# It was expected that for example Gross and star_Gross will appear highly correlated because majority
# of actors played only in one film (according to this dataset). So we can't get any useful information from it

# And from this grafs we can't derive any definitive conclusion about relationship between gross and rating
```

Box Plot of Genres vs. Ratings:



```python
genre_raiting = df[['Genre_list', 'IMDB_Rating']].explode('Genre_list')

plt.figure(figsize=(12, 6))
sns.boxplot(x='Genre_list', y='IMDB_Rating', data=genre_raiting)

median_rating = df['IMDB_Rating'].median()

plt.axhline(y=median_rating, color='r', linestyle='--', label=f'Median Rating: {median_rating:.2f}')

plt.title('Distribution of IMDB Ratings Across Genres')
plt.xlabel('Genre')
plt.ylabel('IMDB Rating')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()

plt.show()
```

```python
genre_counts = df['Genre_list'].explode().value_counts().drop('Drama')

non_drama_df = df[~df['Genre_list'].apply(lambda x: 'Drama' in x)]
nd_genre_counts = non_drama_df['Genre_list'].explode().value_counts()

nd_genre_ratio = (nd_genre_counts / genre_counts).sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=nd_genre_ratio.index, y=nd_genre_ratio.values)

plt.title('Ratio of Non-Drama Films to All Films by Genre')
plt.xlabel('Genre')
plt.ylabel('Ratio')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

Correlation Heatmap:



```python
correlation_matrix = df[['IMDB_Rating', 'Meta_score', 'No_of_Votes', 'Gross', 'Released_Year']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)

plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```

```python
sns.set(style="ticks")

subset_df = df[['No_of_Votes', 'Gross', 'IMDB_Rating']].copy()

g = sns.pairplot(subset_df, vars=['No_of_Votes', 'Gross', 'IMDB_Rating'], 
                 diag_kind='kde', plot_kws={'alpha': 0.6})

g.fig.suptitle('Relationships between Number of Votes, Gross Revenue, and IMDB Rating', y=1.02)

for i, var in enumerate(['Number of Votes', 'Gross Revenue', 'IMDB Rating']):
    g.axes[i, 0].set_ylabel(var)
    g.axes[-1, i].set_xlabel(var)

plt.tight_layout()
plt.show()

correlations = subset_df[['No_of_Votes', 'Gross', 'IMDB_Rating']].corr()
print("Correlations:")
print(correlations)
```

```python
# We can see notable correlation between number of votes and gross. It is pretty obvious that 
# the more film earned the more people watched and rated.
# There is also noticeable correlation between gross and rating. We expect that if people liked the film, 
# they would recommend it to other to watch it. And also it would appear more in media elevating peoples interest.
```
