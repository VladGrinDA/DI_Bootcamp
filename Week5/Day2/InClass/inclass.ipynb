```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample data
data = sns.load_dataset('tips')
sns.histplot(data['total_bill'])
plt.title('Histogram of Total Bill')
plt.show()

```

```python
fig, ax1 = plt.subplots()

sns.histplot(data['total_bill'], bins=20, color='skyblue', ax=ax1)
ax1.set_ylabel('Count')

ax2 = ax1.twinx()
sns.kdeplot(data['total_bill'], color='red', ax=ax2)
ax2.set_ylabel('Density')
plt.title('Histogram and KDE of Total Bill')

plt.show()
```

```python
sns.histplot(data['total_bill'], color='green')
plt.title('Univariate Distribution of Total Bill')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='scatter')
plt.title('Bivariate Distribution of Total Bill and Tips', pad=70)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.histplot(data['total_bill'])
plt.show()
```

```python
sns.set_palette('pastel')
sns.kdeplot(data['total_bill'], fill=True, bw_method='silverman')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
titanic = sns.load_dataset('titanic')

# Creating a bar plot
sns.barplot(x='class', y='survived', data=titanic, hue='sex', errorbar=None)
# Add annotations to show the value on each bar
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height():.2f}', 
                       (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center', 
                       xytext=(0, 9), 
                       textcoords='offset points')

plt.title('Survival Rate by Class')
plt.show()

# Bar Plots are best when you need to compare the quantity of categorical data across different categories.
```

```python
# Creating a bar plot
fig, ax = plt.subplots()

sns.barplot(x='class', y='survived', data=titanic, hue='sex', errorbar=None)
# Add annotations to show the value on each bar
for i in ax.containers:
    ax.bar_label(i, fmt='%.2f')

plt.title('Survival Rate by Class')
plt.show()
```

```python
# Creating a box plot
sns.boxplot(x='class', y='age', data=titanic)
plt.title('Age Distribution by Class')
plt.show()

# Box Plots are ideal for visualizing the distribution of data, making it easy to see the median, quartiles, and outliers.
```

```python
# Creating a violin plot
sns.violinplot(x='class', y='age', data=titanic)
plt.title('Age Distribution by Class')
plt.show()

# Violin Plots are useful when both the distribution of the data and the number of data points are of interest.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading example dataset
data = sns.load_dataset('tips')

# Creating the scatter plot
sns.lmplot(x='total_bill', y='tip', data=data)

# Adding plot title and labels
plt.title('Scatter Plot with Regression Line')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

# Displaying the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading example dataset
data = sns.load_dataset('tips')

# Creating the joint plot
sns.jointplot(x='total_bill', y='tip', data=data, kind='reg')

# Adding a title
plt.subplots_adjust(top=0.9)
plt.suptitle('Joint Plot of Total Bill and Tip')

# Displaying the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
iris = sns.load_dataset('iris')

# Create a pair plot
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])
plt.show()
```

```python
# Calculating the correlation matrix
# We need to exclude non-numeric columns
numeric_iris = iris.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_iris.corr()

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
data = sns.load_dataset("tips")

# Create a facet grid
g = sns.FacetGrid(data, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")

plt.show()
```

```python
# Load the dataset
iris = sns.load_dataset("iris")

# Create a pair grid
g = sns.PairGrid(iris, hue="species")

# Map different plots to different sections of the grid
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_diag(sns.histplot)

plt.show()
```
