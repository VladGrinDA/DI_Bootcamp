```python
from plotnine import ggplot, aes, geom_point
import pandas as pd
import numpy as np

n = 100
x = np.linspace(0, 10, n)
y = np.sin(x) + np.random.normal(0, 0.1, n)

# Sample data
data = pd.DataFrame({'x': x, 'y': y})

# Creating a scatter plot
(ggplot(data, aes(x='x', y='y')) + geom_point())
```

```python
from plotnine import geom_smooth

# Adding a smooth line layer
(ggplot(data, aes(x='x', y='y')) + 
 geom_point() + 
 geom_smooth(method='lm'))
```

```python
import pandas as pd
import numpy as np

from plotnine import ggplot, aes, geom_point, scale_color_manual# Set random seed for reproducibility
np.random.seed(42)# Generate example data

x = np.linspace(1, 10, 200)
y = 2 * x + np.random.normal(0, 2, 200)# Create a categorical variable

category = np.random.choice(['A', 'B', 'C'], size=200)# Create a DataFrame

data = pd.DataFrame({'x': x, 'y': y, 'category': category})# Plot with customized colors

plot = (ggplot(data, aes(x='x', y='y', color='category')) +
        geom_point() +
        scale_color_manual(values=["blue", "green", "red"]))# Display the plot

plot.show()
```

```python
from plotnine import theme_bw

(ggplot(data, aes(x='x', y='y')) + 
 geom_point() + 
 theme_bw())
```

```python
from plotnine import facet_wrap

# Assuming 'category' is a column in your DataFrame
(ggplot(data, aes(x='x', y='y')) + 
 geom_point() + 
 facet_wrap('~category'))
```

```python
from plotnine import geom_histogram

(ggplot(data, aes(x='y')) + 
 geom_histogram(binwidth=1))
```
