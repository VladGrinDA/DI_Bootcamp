Each plot should be well-labeled with appropriate titles and axis labels.

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('iris')
df.head()
```


### 2. Matplotlib Task



```python

plt.figure(figsize=(10, 6))
plt.grid(True)
plt.hist(df['petal_length'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Histogram of Petal Lengths')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()

```

### 3. Seaborn Task

```python

import seaborn as sns

g = sns.pairplot(df, hue="species", height=3, plot_kws={'s': 20})  # Reduced marker size
g.map_lower(sns.kdeplot, levels=4)

plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()


```

### Optional

### 1. Plotnine Task


```python
# Create a scatter plot of sepal length vs. sepal width, color-coded by species.
```

```python
from plotnine import ggplot, aes, geom_point, labs, theme_minimal, scale_color_discrete

plot = (ggplot(df, aes(x='sepal_length', y='sepal_width', color='species')) +
        geom_point(alpha=0.7) + 
        labs(title='Sepal Length vs Sepal Width by Species',
             x='Sepal Length',
             y='Sepal Width') +
        theme_minimal())

plot.show()
```

### 2. Plotly Task


```python
# Create an interactive 3D scatter plot mapping sepal length, sepal width, 
# and petal length, color-coded by species.
```

```python
import plotly.express as px

fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', z='petal_length',
                    color='species', 
                    labels={'sepal_length': 'Sepal Length',
                            'sepal_width': 'Sepal Width',
                            'petal_length': 'Petal Length'},
                    title='Mesurements of Iris',
                    width=800, height=600) 

fig.update_layout(scene_camera=dict(eye=dict(x=1.5, y=1.5, z=0.5)))

fig.show()
```
