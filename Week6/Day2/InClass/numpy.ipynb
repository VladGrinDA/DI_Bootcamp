# Hypothesis Testing

## What is Hypothesis Testing?

Hypothesis testing is a statistical method used to determine the likelihood of a hypothesis being true based on sample data. It involves making an initial assumption (the null hypothesis) and testing this assumption to infer about a population.

## Example of Hypothesis Testing:

Consider a company claims that their new product increases the average duration of customer engagement on their platform from 15 minutes to 20 minutes. To test this claim, a sample of 30 customers is observed.

- **Null Hypothesis (H0):** The average engagement duration is 15 minutes.
- **Alternative Hypothesis (H1):** The average engagement duration is more than 15 minutes.

## Importance and Usage:

- Hypothesis testing is vital in research for validating theories and results.
- Commonly used in fields like science, economics, and medicine to draw conclusions from data.

## How to Conduct Hypothesis Testing:

1. Define Null and Alternative Hypotheses.
2. Choose a significance level (alpha, typically 0.05).
3. Calculate the test statistic and p-value from the sample data.
4. Compare the p-value with the alpha level to accept or reject the Null Hypothesis.

```python
import numpy as np
from scipy import stats

# Sample data (engagement time in minutes)
sample_data = np.random.normal(loc=30, scale=5, size=10)

# Performing a one-sample t-test
t_statistic, p_value = stats.ttest_1samp(sample_data, 15)

# Interpreting the result
alpha = 0.05  # Significance level
print(p_value, t_statistic)
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
import numpy as np
from scipy.stats import skew, kurtosis, lognorm
import matplotlib.pyplot as plt

# Generate a normal distribution (heights in cm)
mean_height = 170
std_dev_height = 10
num_samples = 1000
normal_heights = np.random.normal(mean_height, std_dev_height, num_samples)

# Generate a skewed distribution with high kurtosis
s = 1  # Shape parameter
scale = np.exp(3)  # Scale parameter
skewed_heights = lognorm.rvs(s=s, scale=scale, size=num_samples)

# Calculating skewness and kurtosis for both distributions
normal_skewness = skew(normal_heights)
normal_kurtosis = kurtosis(normal_heights)
skewed_skewness = skew(skewed_heights)
skewed_kurtosis = kurtosis(skewed_heights)

# Print results
print("Normal Distribution:")
print("Skewness:", normal_skewness)
print("Kurtosis:", normal_kurtosis)
print("\nSkewed Distribution:")
print("Skewness:", skewed_skewness)
print("Kurtosis:", skewed_kurtosis)

# Visualize both distributions
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

ax1.hist(normal_heights, bins=30, edgecolor='black')
ax1.set_title('Normal Distribution of Heights')
ax1.set_xlabel('Height (cm)')
ax1.set_ylabel('Frequency')

ax2.hist(skewed_heights, bins=30, edgecolor='black')
ax2.set_title('Skewed Distribution with High Kurtosis')
ax2.set_xlabel('Value')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Example: Using rolling statistics
ts_data = pd.Series(np.random.randn(100), index=pd.date_range('today', periods=100))
rolling_mean = ts_data.rolling(window=5).mean()
print("Rolling Mean:\n", rolling_mean)

# Example: Applying a window function
window_sum = ts_data.rolling(window=5).sum()
print("Window Sum:\n", window_sum)

plt.plot(ts_data.index, rolling_mean)
plt.plot(ts_data.index, window_sum)
plt.plot(ts_data.index, ts_data)
```
