```python
import scipy.stats as stats
import numpy as np

data = [12, 15, 13, 12, 18, 20, 22, 21]
data_ds = stats.describe(data)

for st, val in data_ds._asdict().items():
    print(f'{st}: {val}')
    
print(f'median: {np.median(data)}')
```

Ex 3.

```python
x1 = np.random.normal(loc=0, scale=1, size=100)
x2 = np.random.normal(loc=0.5, scale=2, size=100)

# Perform 2 sample t-test
t_stat, p_val = stats.ttest_ind(x1, x2, equal_var=False)

print(f't-statistic: {t_stat}')
print(f'p-value: {p_val}')

alpha = 0.05

if p_val < alpha:
    print('Difference in means is significant')
else:
    print('Difference in means is not significant')
```

```python
import pandas as pd
from scipy import stats
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

sepal_length_stats = stats.describe(iris_df['sepal length (cm)'])

for st, val in sepal_length_stats._asdict().items():
    print(f'{st}: {val}')
```

Ex 6.

```python
data = np.random.normal(0, 1, 1000)

# Perform a one-sample Kolmogorov-Smirnov test
# This test checks if our data follows a standard normal distribution
ks_stat, p_value = stats.kstest(data, 'norm')

print("Test results:")
print(f"KS Statistic: {ks_stat}")
print(f"p-value: {p_value}")

alpha = 0.05
if p_value > alpha:
    print("The data likely follows a normal distribution.")
else:
    print("The data likely does NOT follow a normal distribution.")
```
