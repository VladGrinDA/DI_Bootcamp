```python
import scipy.stats as stats
```

```python
lambda_ = 3  # average number of calls per hour
poisson_dist = stats.poisson(lambda_)
print("PMF for 3 events:", poisson_dist.pmf(3))
```

```python
# Generate a range of values for x (number of events)
x = range(0, 15)

# Calculate PMF for each x
pmf_values = poisson_dist.pmf(x)

# Calculate CDF for each x
cdf_values = poisson_dist.cdf(x)

# Plotting
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# PMF plot
plt.subplot(1, 2, 1)
plt.bar(x, pmf_values, alpha=0.8, color='b', label='PMF')
plt.title(f'Poisson Distribution (λ = {lambda_})')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.legend()

# CDF plot
plt.subplot(1, 2, 2)
plt.step(x, cdf_values, where='post', label='CDF')
plt.title(f'Poisson Distribution CDF (λ = {lambda_})')
plt.xlabel('Number of Events')
plt.ylabel('Cumulative Probability')
plt.legend()

plt.tight_layout()
plt.show()

# Calculate and print some statistics
mean = poisson_dist.mean()
variance = poisson_dist.var()
std_dev = poisson_dist.std()

print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

```python
import numpy as np
from scipy import stats

# Example data
x = np.array([5, 10, 15, 20, 25])
y = np.array([12, 24, 37, 43, 50])

# Performing linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# The slope and intercept of the regression line
print("Slope:", slope)
print("Intercept:", intercept)

# Plotting the data points and the regression line
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, slope * x + intercept, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()

```

```python
from scipy import stats

# Sample data from three different groups
group1 = [20, 21, 19, 22, 24]
group2 = [28, 30, 27, 26, 29]
group3 = [33, 35, 34, 31, 32]

# Performing one-way ANOVA
f_value, p_value = stats.f_oneway(group1, group2, group3)
print("F-value:", f_value)
print("P-value:", p_value)

# Plotting the data of a one-way ANOVA
plt.figure(figsize=(8, 6))
plt.boxplot([group1, group2, group3], labels=['Group 1', 'Group 2', 'Group 3'])
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('One-way ANOVA')
plt.show()
```
