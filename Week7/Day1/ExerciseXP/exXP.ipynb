```python
def read_file(file_path):
    with open(file_path, 'r') as file:
        contents = file.read()
    return contents
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime

def get_security_alerts_page(year, page_number):
    url = f'https://www.cisa.gov/news-events/cybersecurity-advisories? \
        f%5B0%5D=advisory_type%3A93&f%5B1%5D=release_date_year%3A{year}&page={page_number}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup

def get_security_alerts_count(year):
    page_number = 1
    alerts_count = 0
    while True:
        print(f'Page: {page_number}')
        soup = get_security_alerts_page(year, page_number)
        alerts = soup.find_all(class_='c-view__row')
        if not alerts:
            break
        alerts_count += len(alerts)
        page_number += 1
    
    return alerts_count

get_security_alerts_count(2024)
```

```python
url = 'https://editorial.rottentomatoes.com/guide/popular-movies/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify())
```

```python
movies = soup.find_all(class_='row countdown-item')
print(len(movies))
movies[0]
```

```python
movie = movies[0]
title = movie.find('div', class_="article_movie_title").a.get_text().strip()
year = movie.find('div', class_="article_movie_title").find('span', class_="subtle start-year").get_text()[1:5].strip()
description = movie.find(class_="info synopsis").get_text().strip().split(': ', 1)[1]
title, year, description
```

```python
def get_top_movies(n=10):
    url = 'https://editorial.rottentomatoes.com/guide/popular-movies/'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    movies = soup.find_all(class_='row countdown-item')
    
    top_movies = []
    for movie in movies[:n]:
        title = movie.find('div', class_="article_movie_title").a.get_text().strip()
        year = movie.find('div', class_="article_movie_title").find('span', class_="subtle start-year").get_text()[1:5].strip()
        description = movie.find(class_="info synopsis").get_text().strip().split(': ', 1)[1]
        top_movies.append({'title': title, 'year': year, 'description': description})
    
    return top_movies

get_top_movies()
```
