```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from pprint import pprint  # To tidy up
from bs4 import BeautifulSoup
import time

options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run Chrome in headless mode
options.add_argument("--no-sandbox")  # Bypass OS security model
options.add_argument("--disable-dev-shm-usage")  # Overcome limited resource problems
```

```python
driver = webdriver.Chrome(options=options)
url = "https://www.rottentomatoes.com/browse/movies_at_home/critics:certified_fresh"
driver.get(url)



try:
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "div.flex-container "))
    )
except Exception as e:
    print("Error loading page:", e)
    driver.quit()
    exit()

# n = 2
# for i in range(n):
#     load_more_button = driver.find_element((By.CSS_SELECTOR, 'button[data-qa="dlp-load-more-button"]'))
#     load_more_button.click()
#     time.sleep(2)    

html_content = driver.page_source
soup = BeautifulSoup(html_content, 'html.parser')

movies = soup.select("div.flex-container ")
movies[0]

driver.quit()
```

```python
from datetime import datetime

movie = movies[0]

title = movie.select_one("span.p--small").get_text(strip=True)
critic_score = movie.find('rt-text', slot='criticsScore').get_text(strip=True).replace('%', '')
audience_score = movie.find('rt-text', slot='audienceScore').get_text(strip=True).replace('%', '')

streaming_date = movie.find('span', class_='smaller', attrs={'data-qa': 'discovery-media-list-item-start-date'}).get_text(strip=True)
streaming_date = datetime.strptime(streaming_date.replace('Streaming ', ''), '%b %d, %Y').strftime('%Y-%m-%d')

title, critic_score, audience_score, streaming_date
```

```python
movie
```

```python
def get_movie_data(movie):
    title = movie.select_one("span.p--small").get_text(strip=True)
    
    critic_score = movie.find('rt-text', slot='criticsScore')
    if critic_score:
        critic_score = critic_score.get_text(strip=True).replace('%', '')
        
    audience_score = movie.find('rt-text', slot='audienceScore')
    if audience_score:
        audience_score = audience_score.get_text(strip=True).replace('%', '')
        
    streaming_date = movie.find('span', class_='smaller', attrs={'data-qa': 'discovery-media-list-item-start-date'}).get_text(strip=True)
    streaming_date = datetime.strptime(streaming_date.replace('Streaming ', ''), '%b %d, %Y').strftime('%Y-%m-%d')
    
    return title, critic_score, audience_score, streaming_date

movies_data = []
for i, movie in enumerate(movies):
    # print(f"{i+1} of {len(movies)}")
    title, critic_score, audience_score, streaming_date = get_movie_data(movie)
    movies_data.append({'title': title, 'critic_score': critic_score, 'audience_score': audience_score, 'streaming_date': streaming_date})

pprint(movies_data)
```

```python

# url = "https://www.chathamhouse.org/search/content_format/Article"
# Almost every news site had a protection from bots, so I used this one
url = "https://www.ynetnews.com/category/3082"
driver.get(url)

pprint(driver.page_source[0:1000])
```

```python
wait = WebDriverWait(driver, 10)
nav_list = wait.until(EC.visibility_of_element_located((
    By.CSS_SELECTOR, 
    'div.navList'
)))

# Distinct navigation elements had a different id every time the page was loaded
nav_elements = nav_list.find_elements(By.XPATH, "./*")
second_child_element = nav_elements[1] # The second one is for news section
link = second_child_element.find_element(By.TAG_NAME, "a").get_attribute("href")
driver.get(link)

pprint(driver.page_source[:1000])
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
# with open("soup_output.html", "w", encoding="utf-8") as file:
#     file.write(str(soup))
```

```python
articles = soup.find_all('div', class_='textDiv')
articles[0]
```

```python
from datetime import datetime

article = articles[0]

title = article.find('div', class_='slotTitle').get_text(strip=True)
subtitle = article.find('div', class_='slotSubTitle').get_text(strip=True)
author = article.find('span', class_='authorInfo').get_text(strip=True)
date = article.find('span', class_='dateView').get_text(strip=True)

date = datetime.strptime(date, '%H:%M | %m.%d.%y')
date_str = date.strftime('%Y-%m-%d %H:%M')


title, subtitle, author, date, date_str
```

```python
articles_data = []
for article in articles:
    title = article.find('div', class_='slotTitle').get_text(strip=True)
    subtitle = article.find('div', class_='slotSubTitle').get_text(strip=True)
    author = article.find('span', class_='authorInfo').get_text(strip=True)
    date = article.find('span', class_='dateView').get_text(strip=True)
    
    date = datetime.strptime(date, '%H:%M | %m.%d.%y')
    date_str = date.strftime('%Y-%m-%d %H:%M')
    
    articles_data.append({'title': title, 'subtitle': subtitle, 'author': author, 'date': date, 'date_str': date_str})

pprint(articles_data)

driver.quit()
```

```python
import pandas as pd

df = pd.DataFrame(articles_data)
df.head()
```

```python
# All the news are from current month so let's catagorize them by day
df['date'] = pd.to_datetime(df['date'])
df['day'] = df['date'].dt.day
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year

df.head()
```
